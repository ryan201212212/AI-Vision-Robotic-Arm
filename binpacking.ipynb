```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots()

items_floor1 = [
    {'name': 'water', 'coordinates': (176.97, 226.92), 'width': 216.97, 'height': 77.9},
    {'name': 'egg', 'coordinates': (139.56, 375.84), 'width': 134.95, 'height': 98},
    {'name': 'milk', 'coordinates': (300.16, 413.95), 'width': 77.75, 'height': 95 },
]

spaceWidth = 230
spaceHeight = 180
maxHeight = 0

nextPos = {'x': 0, 'y': 0}

colors_floor1 = ['red', 'green', 'blue']
for i, item in enumerate(items_floor1):
    if nextPos['x'] + item['width'] > spaceWidth:
        nextPos['x'] = 0
        nextPos['y'] += maxHeight
        maxHeight = 0

    if nextPos['y'] + item['height'] <= spaceHeight:
        item['x'] = nextPos['x']
        item['y'] = nextPos['y']
        nextPos['x'] += item['width'] + 10
        maxHeight = max(maxHeight, item['height'])

        rect = patches.Rectangle((item['x'], item['y']), item['width'], item['height'],
                                 linewidth=1, edgecolor='r', facecolor=colors_floor1[i], label=item['name'])
        ax.add_patch(rect)
        plt.text(item['x'], item['y'], item['name'], fontsize=8, weight='bold')

plt.xlim(0, spaceWidth)
plt.ylim(0, spaceHeight)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.show()
```
